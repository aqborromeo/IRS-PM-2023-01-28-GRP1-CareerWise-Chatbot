```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pprint
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors
from gensim.test.utils import datapath
```

```python
import re

def preprocess(text):
    text = text.lower()
    text=re.sub('[^a-z]+', ' ',text)
    return text
```

```python
# load w2v from pre-built Google data
w2v = word2vec.Word2Vec()
# download bin.gz from: https://code.google.com/archive/p/word2vec/
w2v = KeyedVectors.load_word2vec_format(datapath("/home/e1113633/project/GoogleNews-vectors-negative300.bin"), binary=True)

#w2v_vocab = set(w2v.vocab)

print("Loaded {} words in vocabulary".format(len(w2v)))
```

```python
# read careers future dataset
cf_df = pd.read_csv("/home/e1113633/project/cf_data.csv")
cf_df = cf_df[pd.notnull(cf_df['job_title'])]
cf_df['job_title'].apply(lambda x: len(x.split(' '))).sum()

cf_df.head(10)
```

```python
# read onet dataset
df = pd.read_csv('/home/e1113633/project/onet_ssoc_riasec.csv')
df.head(10)
```

```python
onet_job = preprocess(df['ssoc_job_title'][1]).split()
cf_job = preprocess(cf_df['job_title'][1]).split()
print(onet_job)
print(cf_job)
#w2v.n_similarity(['ambassador', 'government'], ['services', 'ambassador', 'banquet'])
w2v.n_similarity(onet_job, cf_job)
```

```python
# construct a new DataFrame 
new_df = pd.DataFrame(data=None, index=None, columns=df.columns, dtype=None, copy=None)

new_df["cf_id"] = ''
new_df["cf_job_title"] = ''
new_df["min_salary"] = ''
new_df["max_salary"] = ''

new_df
```

```python
from operator import itemgetter
# find matching careers future job title for each onet job title
max_match = 4
i = 0
for onet_job_title in df['ssoc_job_title']:
    j=0
    threshold = 0.7
    cf_job_ids = []
    for cf_job_title in cf_df['job_title']:
        onet_job = preprocess(onet_job_title).split()
        cf_job = preprocess(cf_job_title).split()
        if (len(onet_job) == 0) or (len(cf_job) == 0):
            continue
            
        score = w2v.n_similarity(onet_job, cf_job)
        if score > threshold:
            # if found a match, append to new dataframe
            if (len(cf_job_ids) == max_match):
                cf_job_ids = sorted(cf_job_ids, key=itemgetter(1))
                threshold = cf_job_ids[0][1]
            if (len(cf_job_ids) > max_match):
                # if score is higher than any of the existing one, replace the lowest score
                cf_job_ids = sorted(cf_job_ids, key=itemgetter(1))
                cf_job_ids[0] = [j, score]
                threshold = cf_job_ids[0][1]   
            else:
                cf_job_ids.append([j, score])
        j+=1  
        
    # add to job list
    for cf_job_id in cf_job_ids:
        new_df.loc[len(new_df)] = df.loc[i].tolist() + cf_df[['cf_id', 'job_title', 'min salary', 'max salary']].loc[cf_job_id[0]].tolist()

    i+=1
    if (i % 1000 == 0):
        print("processed ", i)
    #if (i > 20):
        #break
#new_df.head(20)  
```

```python
new_df.to_csv("cf_onet.csv",  encoding='utf-8')
```
